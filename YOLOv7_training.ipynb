# –ê–Ω–∞–ª–∏–∑ –æ–±—É—á–µ–Ω–∏—è

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

s = ['train', 'train', 'train', 'Precision', 'mAP@0.5', 'val', 'val', 'val', 'Recall', 'mAP@0.5:0.95']  # legends
t = ['Box loss', 'Objectness loss', 'Classification loss', 'P-R', 'mAP-F1']  # titles
f = '/content/gdrive/MyDrive/yolov7_run/yolov7/runs/train/yolov7-custom2/results.txt'
start, stop = 0, 99

results = np.loadtxt(f, usecols=[2, 3, 4, 8, 9, 12, 13, 14, 10, 11], ndmin=2).T
n = results.shape[1]  # number of rows
x = range(start, min(stop, n) if stop else n)


plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x, results[0, x], marker='.', label=s[0])
plt.plot(x, results[5, x], marker='.', label=s[5])
plt.title(t[0])
plt.xlabel('–ù–æ–º–µ—Ä —ç–ø–æ—Ö–∏')
plt.ylabel('Box_loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, results[1, x], marker='.', label=s[1])
plt.plot(x, results[6, x], marker='.', label=s[6])
plt.title(t[1])
plt.xlabel('–ù–æ–º–µ—Ä —ç–ø–æ—Ö–∏')
plt.ylabel('Objectness loss')
plt.legend()
plt.show()


plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(14, 3.5), tight_layout=True)
ax = ax.ravel()
for i in range(5):
    for j in [i, i + 5]:
        y = results[j, x]
        ax[i].plot(x, y, marker='.', label=s[j])
        # y_smooth = butter_lowpass_filtfilt(y)
        # ax[i].plot(x, np.gradient(y_smooth), marker='.', label=s[j])

    ax[i].set_title(t[i])
    ax[i].legend()
    ax[i].set_ylabel(f) if i == 0 else None  # add filename
```
