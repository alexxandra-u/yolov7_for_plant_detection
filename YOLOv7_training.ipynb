```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

s = ['train', 'train', 'train', 'Precision', 'mAP@0.5', 'val', 'val', 'val', 'Recall', 'mAP@0.5:0.95']  # legends
t = ['Box loss', 'Objectness loss', 'Classification loss', 'P-R', 'mAP-F1']  # titles
f = '/content/gdrive/MyDrive/yolov7_run/yolov7/runs/train/yolov7-custom2/results.txt'
start, stop = 0, 99

results = np.loadtxt(f, usecols=[2, 3, 4, 8, 9, 12, 13, 14, 10, 11], ndmin=2).T
n = results.shape[1]  # number of rows
x = range(start, min(stop, n) if stop else n)


plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x, results[0, x], marker='.', label=s[0])
plt.plot(x, results[5, x], marker='.', label=s[5])
plt.title(t[0])
plt.xlabel('ÐÐ¾Ð¼ÐµÑ€ ÑÐ¿Ð¾Ñ…Ð¸')
plt.ylabel('Box_loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, results[1, x], marker='.', label=s[1])
plt.plot(x, results[6, x], marker='.', label=s[6])
plt.title(t[1])
plt.xlabel('ÐÐ¾Ð¼ÐµÑ€ ÑÐ¿Ð¾Ñ…Ð¸')
plt.ylabel('Objectness loss')
plt.legend()
plt.show()


plt.show()
```
